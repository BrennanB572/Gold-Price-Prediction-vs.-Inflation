```python
# Importing Modules
import pandas as pd
import numpy as np
```

```python
# Referencing the China Averages File that has China to US Gold Currency Conversions
china_avgs = ('Resources/CNY2US.csv')

# Reading the data then displaying the top 5 results
df = pd.read_csv(china_avgs)
df.head()
```

```python
# Organizing the China Averages Files and creating a new DataFrame that groups by year and outputting the new results into a new DataFrame
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
columns_to_average = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]
average_annual_data = pd.DataFrame()
```

```python
# Looping through columns in new DataFrame "average_annual_data" and then grouping by year and averaging each column
for column in columns_to_average:
    average_annual_data[column + '_Avg'] = df.groupby('Year')[column].mean()
    print(average_annual_data)
```

```python
# File export into => China_Averages.csv
average_annual_data.to_csv('Resources/CHINA_averages.csv', header=True)
```
