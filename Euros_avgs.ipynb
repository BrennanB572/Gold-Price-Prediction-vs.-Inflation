```python
#Importing dependencies
import pandas as pd
import numpy as np
```

```python
#Reading csv
euros_avgs = ('Resources/EURO2US.csv')
df = pd.read_csv(euros_avgs)
df.head()
```

```python
#Pulling the years from the date time
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
columns_to_average = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]
average_annual_data = pd.DataFrame()
```

```python
#Creating a loop to pull averages per year
for column in columns_to_average:
    average_annual_data[column + '_Avg'] = df.groupby('Year')[column].mean()
    print(average_annual_data)
```

```python
#Exporting new csv file
average_annual_data.to_csv('Resources/EUROS_averages.csv', header=True)
```
