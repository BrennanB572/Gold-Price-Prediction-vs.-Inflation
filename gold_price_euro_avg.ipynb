```python
#Import dependencies
import pandas as pd
import numpy as np
```

```python
#Reading csv files
average_annual_gold_prices_df = pd.read_csv('Resources/average_annual_gold_prices.csv')
average_annual_gold_prices_df
```

```python
EUROS_averages_df = pd.read_csv('Resources/EUROS_averages.csv')
EUROS_averages_df
```

```python
#Merging into one file to prepare for looping
merged_df = pd.merge(average_annual_gold_prices_df, EUROS_averages_df, how='left', on = ['Year', 'Year'])
merged_df
```

```python
#Removing NaN values and empty rows
nan_value = float("NaN") 
merged_df.replace(0, nan_value, inplace=True) 
merged_df['Vol._Avg', 'Change %_Avg', 'Volume_Avg'] = np.nan 
merged_df.dropna(how='all', axis=1, inplace=True)
merged_df_clean = merged_df
merged_df_clean
```

```python
#Converting the Price_Avg (USD) to Euros by dividing by Close_Avg(Conversion Rate)
merged_df_clean['Gold_Price_Euros'] = merged_df_clean['Price_Avg'].div(merged_df_clean['Close_Avg'], axis=0)
merged_df_clean
```

```python
#Creating smaller dataframe with euro gold prices per year (2013-2022)
gold_price_euros_df = merged_df_clean[['Year', 'Gold_Price_Euros']]
gold_price_euros_df = gold_price_euros_df.set_index('Year')
gold_price_euros_df
```

```python
#Saving to csv file
gold_price_euros_df.to_csv('Resources/Gold_Price_Euro_Conversion(2013-2022).csv', header=True)
```
